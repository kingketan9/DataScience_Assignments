2.Draw pie chart

```python
from matplotlib import pyplot as plt
import numpy as np

Languages = ['JAVA', 'PYTHON', 'PHP',
		'JAVASCRIPT', 'C#', 'C++']

popularity = [22.2, 17.6, 8.8, 8.0, 7.7, 6.7]

fig = plt.figure(figsize =(10, 7))
plt.pie(popularity, labels = Languages)

plt.show()
```
